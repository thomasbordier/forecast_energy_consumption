# TRAITEMENT FICHIER METEO

```python
df_weather = pd.read_csv("../raw_data/weather_day_2013_06132022.csv")
df_weather
```

```python
df_weather.drop(['ALLSKY_KT','ALLSKY_SFC_PAR_TOT','CLRSKY_SFC_PAR_TOT','ALLSKY_SFC_SW_DWN','CLRSKY_SFC_SW_DWN',], axis=1, inplace = True)
```

```python
df_weather.drop(df_weather.tail(2).index, inplace = True) 
```

```python
df_weather
```

```python
df_weather.rename(columns = {'MO': 'MONTH','DY': 'DAY'}, inplace = True)
```

```python
# CREATION CHAMPS DATE

df_weather['Date'] = pd.to_datetime(df_weather[['YEAR','MONTH', 'DAY']], errors = 'coerce')
```

```python
df_weather['sin_month'] = np.sin(2*np.pi*df_weather.MONTH/12)
df_weather['cos_month'] = np.cos(2*np.pi*df_weather.MONTH/12)
```

```python
df_weather['week_day'] = df_weather['Date'].apply(lambda x : x.weekday())
df_weather['sin_day'] = np.sin(2*np.pi*df_weather.week_day/7)
df_weather['cos_day'] = np.cos(2*np.pi*df_weather.week_day/7)
```

```python
df_weather.head()
```

```python
df_weather.reset_index(drop=True)
df_weather.to_csv('../raw_data/df_weather_clean.csv')
```

# VACANCES SCOLAIRE & JOURS (COS, SIN)

```python
df_holiday = pd.read_csv("../raw_data/data_holiday.csv")
df_holiday = df_holiday.rename(columns={'date': 'Date','vacances_zone_b':'school_holidays'})
```

```python
df_holiday = df_holiday.drop(["vacances_zone_a", "vacances_zone_c",'nom_vacances'], axis=1)
```

```python
df_holiday['school_holidays']=df_holiday['school_holidays'].replace({False:0,True:1})
df_holiday['Date'] = pd.to_datetime(df_holiday['Date'], infer_datetime_format=True)
df_holiday
```

```python
def season_of_date(date):
    year = str(date.year)
    seasons = {'spring': pd.date_range(start='21/03/'+year, end='20/06/'+year),
               'summer': pd.date_range(start='21/06/'+year, end='22/09/'+year),
               'autumn': pd.date_range(start='23/09/'+year, end='20/12/'+year)}
    if date in seasons['spring']:
        return 2 #'spring'
    if date in seasons['summer']:
        return 3 #'summer'
    if date in seasons['autumn']:
        return 0 #'autumn'
    else:
        return 1 #'winter'

df_holiday['season'] = df_holiday['Date'].apply(season_of_date)
```

```python
df_holiday['num_day'] = df_holiday['Date'].apply(lambda x : x.timetuple().tm_yday)
#df_ferie['sin_num_day'] = np.sin(2*np.pi*df_weather.week_day/365)
#df_ferie['cos_num_day'] = np.cos(2*np.pi*df_weather.week_day/365)
df_holiday['sin365_1']=np.sin(2*np.pi*df_holiday.num_day/365.25)
df_holiday['cos365_1']=np.cos(2*np.pi*df_holiday.num_day/365.25)
df_holiday['sin365_2']=np.sin(4*np.pi*df_holiday.num_day/365.25)
df_holiday['cos365_2']=np.cos(4*np.pi*df_holiday.num_day/365.25)
df_holiday['sin365_3']=np.sin(8*np.pi*df_holiday.num_day/365.25)
df_holiday['cos365_3']=np.cos(8*np.pi*df_holiday.num_day/365.25)
df_holiday
```
