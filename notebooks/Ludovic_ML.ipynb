```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

feat_numerical = sorted(X.select_dtypes(include=["int64", "float64"]).columns)

preproc_MinMaxScaler = make_pipeline(MinMaxScaler())


preproc_StandardScaler = make_pipeline(StandardScaler())
    

preproc_RobustScaler = make_pipeline( RobustScaler())
```

```python
min_max = ['week_day']
stand = ['T2MDEW','T2M_RANGE']
robus = ['T2M', 'T2MWET', 'TS', 'T2M_MAX', 'T2M_MIN','QV2M', 'RH2M', 'PRECTOTCORR', 'PS', 'WS10M', 'WS50M', 'sin_month', 'cos_month', 'sin_day', 'cos_day']

preproc = make_column_transformer(
    (preproc_MinMaxScaler, min_max),
    (preproc_StandardScaler, stand),
    (preproc_RobustScaler, robus),
    remainder="drop")

preproc
```

```python
pd.DataFrame(preproc.fit_transform(X,y)).head()
```

# AdaBoosted Trees à Sklearn

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt
from tempfile import mkdtemp
from sklearn.ensemble import GradientBoostingRegressor
cachedir = mkdtemp()

model_ada = AdaBoostRegressor(
    base_estimator = DecisionTreeRegressor(max_depth=3),
    n_estimators=50)

cv_results = cross_validate(model_ada, X, y, scoring = "neg_mean_absolute_percentage_error", cv=5)
```

```python
cv_results['test_score'].mean()
```

```python
pipe_adaboost = make_pipeline(preproc, model_ada, memory=cachedir)
score = cross_val_score(pipe_adaboost, X, y, cv=5, scoring='neg_mean_absolute_percentage_error')
score.mean()
```

```python
pipe_adaboost.get_params()
```

```python
param = { 'n_estimators':[575, 585], 'learning_rate':[ 1, 1.1, 1.2], 'loss':['linear', 'square','exponential'] } 
#adaRegressor = AdaBoostRegressor(random_state = 10)

grid_ada = GridSearchCV(model_ada, param, cv = 5, n_jobs = -1, verbose=1, scoring='neg_mean_absolute_percentage_error') 
grid_ada.fit(X, y) 
```

```python
grid_ada.best_params_
```

```python
grid_ada.best_score_
```

# GradientBoostingRegressor

```python
model = GradientBoostingRegressor(n_estimators=100, verbose=0)
pipe_gb = make_pipeline(preproc, model, memory=cachedir)
score = cross_val_score(pipe_gb, X, y, cv=5, scoring='neg_mean_absolute_percentage_error')
print(score.std())
score.mean()
```

```python
parameters_GBR = {'learning_rate': [0.011, 0.01,0.02],
                  'subsample'    : [0.9, 0.5, 0.2, 0.1],
                  'n_estimators' : [1250, 1500, 1750],
                  'max_depth'    : [3,4,5]
                 }

GBR = GradientBoostingRegressor()
```

```python
grid_GBR = GridSearchCV(estimator=GBR, param_grid = parameters_GBR, cv = 5, n_jobs=-1, scoring='neg_mean_absolute_percentage_error')
```

```python
grid_GBR.fit(X, y)
```

```python
grid_GBR.best_params_
```

```python
grid_GBR.best_score_
```
